```python
import os
import shutil

src_dir = "../dataset_balanced"
dst_dir = "dataset_new_balanced"

src_images = os.path.join(src_dir, "images")
src_labels = os.path.join(src_dir, "labels")

dst_images = os.path.join(dst_dir, "images")
dst_labels = os.path.join(dst_dir, "labels")

os.makedirs(dst_images, exist_ok=True)
os.makedirs(dst_labels, exist_ok=True)

#keep classes:
target_classes = {2351, 2392, 2991, 3031, 5121, 5953, 6714}

for label_file in os.listdir(src_labels):
    if not label_file.endswith(".txt"):
        continue

    src_label_path = os.path.join(src_labels, label_file)

    with open(src_label_path, "r") as f:
        lines = f.readlines()

    filtered_lines = []
    for line in lines:
        parts = line.strip().split()
        if not parts:
            continue
        try:
            class_id = int(parts[0])
        except ValueError:
            continue
        if class_id in target_classes:
            filtered_lines.append(line)

    #write to new folder
    if filtered_lines:
        dst_label_path = os.path.join(dst_labels, label_file)
        with open(dst_label_path, "w") as f:
            f.writelines(filtered_lines)

        # kopieer de bijbehorende image
        image_name = os.path.splitext(label_file)[0] + ".jpg"
        src_image_path = os.path.join(src_images, image_name)
        dst_image_path = os.path.join(dst_images, image_name)

        if os.path.exists(src_image_path):
            shutil.copy(src_image_path, dst_image_path)
        else:
            print(f"Image not found for label: {label_file}")

print("New, to:", dst_dir)
```

```python
import os
from collections import defaultdict

dataset_dir = "../dataset_new_clean" 

labels_dir = os.path.join(dataset_dir, "labels")

#counters
object_count = defaultdict(int)   
image_count = defaultdict(set)    

#loop
for label_file in os.listdir(labels_dir):
    if not label_file.endswith(".txt"):
        continue

    image_name = os.path.splitext(label_file)[0]  
    label_path = os.path.join(labels_dir, label_file)

    with open(label_path, "r") as f:
        lines = f.readlines()

    for line in lines:
        parts = line.strip().split()
        if not parts:
            continue
        try:
            class_id = int(parts[0])
        except ValueError:
            continue

        object_count[class_id] += 1
        image_count[class_id].add(image_name)

# maak overzicht
print(f"{'Class ID':>8} | {'# Objects':>10} | {'# Images':>10}")
print("-" * 34)
for class_id in sorted(object_count.keys()):
    print(f"{class_id:>8} | {object_count[class_id]:>10} | {len(image_count[class_id]):>10}")
```

```python
import os
import shutil
import random
from collections import defaultdict

# instellingen
src_dir = "../dataset_new_balanced"
dst_dir = "dataset_yolo"

splits = {
    "train": 0.7,  # 70%
    "val": 0.2,    # 20%
    "test": 0.1    # 10%
}

# maak directories aan
for split in splits.keys():
    os.makedirs(os.path.join(dst_dir, split, "images"), exist_ok=True)
    os.makedirs(os.path.join(dst_dir, split, "labels"), exist_ok=True)

src_images = os.path.join(src_dir, "images")
src_labels = os.path.join(src_dir, "labels")

# stap 1: alle labels inlezen per class
class_to_files = defaultdict(list)

for label_file in os.listdir(src_labels):
    if not label_file.endswith(".txt"):
        continue
    path = os.path.join(src_labels, label_file)
    with open(path, "r") as f:
        lines = f.readlines()
    classes_in_file = set(int(l.split()[0]) for l in lines if l.strip())
    for c in classes_in_file:
        class_to_files[c].append(label_file)

# stap 2: maak een evenwichtige verdeling per class
assigned_files = set()
split_assignments = {"train": set(), "val": set(), "test": set()}

for c, files in class_to_files.items():
    random.shuffle(files)
    n = len(files)
    n_train = int(n * splits["train"])
    n_val = int(n * splits["val"])
    n_test = n - n_train - n_val

    split_assignments["train"].update(files[:n_train])
    split_assignments["val"].update(files[n_train:n_train + n_val])
    split_assignments["test"].update(files[n_train + n_val:])

# stap 3: bestanden kopiëren naar juiste mappen
for split, files in split_assignments.items():
    for label_file in files:
        label_src = os.path.join(src_labels, label_file)
        img_name = os.path.splitext(label_file)[0] + ".jpg"
        img_src = os.path.join(src_images, img_name)

        label_dst = os.path.join(dst_dir, split, "labels", label_file)
        img_dst = os.path.join(dst_dir, split, "images", img_name)

        if os.path.exists(label_src):
            shutil.copy(label_src, label_dst)
        if os.path.exists(img_src):
            shutil.copy(img_src, img_dst)

print("✅ YOLO dataset gemaakt in:", dst_dir)
```

```python
from collections import Counter

for split in splits.keys():
    split_labels_dir = os.path.join(dst_dir, split, "labels")
    class_counter = Counter()
    for label_file in os.listdir(split_labels_dir):
        if not label_file.endswith(".txt"):
            continue
        with open(os.path.join(split_labels_dir, label_file), "r") as f:
            for line in f:
                parts = line.strip().split()
                if parts:
                    try:
                        class_id = int(parts[0])
                        class_counter[class_id] += 1
                    except ValueError:
                        continue
    print(f"\nSplit: {split}")
    for cid, count in sorted(class_counter.items()):
        print(f"  Class {cid}: {count} objects")
```
